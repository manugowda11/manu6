```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,5,3,4],[1,3,6,8])
#plt.plot([1,2,3,4][1,8,27,64])
plt.title('cube plot',color='g')
plt.xlabel('NUMBERS N')
plt.ylabel('N**3')
plt.show
plt.annotate('cube result',(1,64))
plt.axis([0,10,0,90])
plt.figure(figsize=(12,7))
```

```python
plt.plot([1,2,3,4],[2,4,6,8],'ro')
plt.title('cube plot',color='g')
plt.xlabel('NUMBERS N')
plt.ylabel('N**3')
plt.show
plt.annotate('cube result',(1,64))
plt.axis([0,10,0,90])
plt.figure(figsize=(12,7))
```

```python
plt.figure(figsize=(12,7))
plt.bar([1,2,3,4],[2,4,6,8])
plt.title('bar plot',color='green')
plt.xlabel('NUMBERS N')
plt.ylabel('N**3')
plt.show
plt.annotate('bar result',(1,64))
plt.axis([0,10,0,90])
plt.figure(figsize=(12,7))
```
